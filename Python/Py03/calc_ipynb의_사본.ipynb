```python
import math

try:
    num1 = int(input('첫 번째 정수를 입력하세요: '))
    num2 = int(input('두 번째 정수를 입력하세요: '))
    operator = input('연산자를 입력하세요 (+, -, *, /, **): ')

    allowed = {'+', '-', '*', '/', '**'}

                                                           # 1. 연산자 유효성 검사
    if operator not in allowed:
        print('오류: 잘못된 연산자입니다.')
    else:
                                                            # 2. 실제 연산 수행 (eval을 쓰거나 if문으로 분기)
        if operator == '+':
          result = num1 + num2
        elif operator == '-':
          result = num1 - num2
        elif operator == '*':
           result = num1 * num2
        elif operator == '/':
          result = num1 / num2          # ZeroDivisionError 발생 가능
        elif operator == '**':
          result = num1 ** num2

        print(f'결과: {result}')

except ValueError:
    print('잘못된 입력입니다. 정수를 입력해주세요.')
except ZeroDivisionError:
    print('오류: 0으로 나눌 수 없습니다.')
except Exception as e:
    print(f'예상치 못한 오류가 발생했습니다: {e}')
finally:
   print('연산 프로세스가 종료되었습니다. 계속 실행하시겠습니까?', )
  #again = input('계속 실행하시겠습니까? (y/n): ').lower()
   # if again == 'n':
       # print('프로그램을 종료합니다.')
        #break                                             # 3. 루프 탈출
```
