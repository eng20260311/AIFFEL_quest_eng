```python
menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]
class Kiosk:
    def __init__(self):
        self.menu = menu
        self.price = price
        self.ordered = []

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        print()  # 줄 바꿈
        n = 0
        while n < 1 or len(menu) < n:
            n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력

            # 메뉴판에 있는 음료 번호일때
            if 1 <= n & n <= len(menu):
                self.price_sum = self.price[n-1]  # 선택 음료의 가격

                print(self.menu[n-1], ' : ', self.price[n-1] , '원')  # 메뉴
                self.ordered.append(n-1)
             # 메뉴판에 없는 번호일 때
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")
                # 음료 온도 물어보기
        t = 0  # 기본값을 넣어주고
        while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
            t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))

            if t==1:
                self.temp='Hot'
            elif t==2:
                self.temp='Ice'
            else:
                print("1과 2 중 하나를 입력하세요.\n")

        print('주문 음료', self.temp, self.menu[n-1], ' : ', self.price[n-1], '원')  # 온도 속성을 추가한 주문 결과 출력

        question=int(input("추가 주문하시겠습니까? 추가 주문하시려면 0(완료) 또는 1(추가주문)을 입력하십시오.: "))
        if question == 1:
            self.menu_select()
    def info(self):
      totalPrice = 0
      print()
      print("주문하신 내역은")
      for t in self.ordered:
        orderdMenu = self.menu[t]
        orderdPrice = self.price[t]
        totalPrice += orderdPrice
        print(f"{orderdMenu} : {orderdPrice}")
      print(f"총 금액은 {totalPrice}원 입니다")

def pay(self):
    print()
    while self.pay != 1 and self.pay != 2:
       self.pay = int(input("결제방법을 입력하세요. cash는 1번, card는 2번을 입력하세요 : "))
    if self.pay== 1:
        print('직원을 호출하겠습니다')
    elif self.pay== 2:
        print('IC칩 방향에 맞게 카드를 꽂아주세요')
    else :
        print('다시 결제를 시도해 주세요')


def table(self):
  print()
  print("\n" + "="*30)
  print("       [ 주문 영수증 ]")
  print("-"*30)

  totalPrice = 0
  for t in self.ordered:
    orderdMenu = self.menu[t]
    orderdPrice = self.price[t]
    totalPrice += orderdPrice
    print(f" 선택 메뉴 : {orderdPrice} {orderdMenu}")
  print(f" 결제 금액 : {totalPrice:,}원")
  print("-"*30)
  print("   이용해 주셔서 감사합니다!")
  print("="*30 + "\n")





Kiosk.pay = pay
Kiosk.table = table
a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.info()
a.table() # 영수증





```

```python
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')

        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        base,height=map(int, input('밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>>').split(','))
        area = base * height
        result = '평행사변형의 넓이는 : ' + str(area)
        return result

    def trape(self):
      upper_base,lower_base, height=map(int, input('윗변, 아랫변, 높이를 입력하세요. 예시 : 윗변,아랫변,높이\n >>>').split(','))
      area = (upper_base+lower_base)*height/2
      result = '사다리꼴의 넓이는: ' + str(area)
      return result

a = Square()
print()
print(a.rect())
print()
print(a.par())
print()
print(a.trape())
```
